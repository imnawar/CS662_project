```python
from faker.providers import BaseProvider
from faker import Faker
import random



class SaudiRegionProvider(BaseProvider):
    def saudi_region(self):
        saudi_regions = [
            "Riyadh", "Makkah", "Madinah", "Eastern Province",
            "Asir", "Tabuk", "Hail", "Qassim", "Najran",
            "Jazan", "Al-Baha", "Al-Jouf", "Northern Borders"
        ]
        return self.random_element(saudi_regions)


# Initialize Faker
fake = Faker('ar')

fake.add_provider(SaudiRegionProvider)


# Function to generate a random student profile
def generate_student_profile():
    student_profile = {
        "student_name": "",
        "studnet_region": "",
        "student_level": 0,
        "Tahsili_target": 0,
        "Qudurat_target": 0,
        "student_courses": [],
        "previous_Qudurat_grade": {"math": 0, "arabic": 0, "all": 0},
        "previous_Tahsili_grade": 0,
        "student_grades": {}
    }
    student_profile['student_name'] = fake.first_name() + " " + fake.last_name()
    student_profile['studnet_region'] = fake.saudi_region()
    student_profile['student_level'] = random.randint(1, 9)
    student_profile['Tahsili_target'] = random.choice([0, 1]) if student_profile['student_level']>6 else 0
    student_profile['Qudurat_target'] = random.choice([0, 1])
    student_profile['student_courses'] = [course for course in df[df['course_level']<student_profile['student_level']]['course_id'].tolist()]
    student_profile['previous_Qudurat_grade']= {
                "math": random.randint(40, 100),
                "arabic": random.randint(40, 100),
                "all": 0
            }
    student_profile['previous_Qudurat_grade']['all'] = (student_profile['previous_Qudurat_grade']['math'] + student_profile['previous_Qudurat_grade']['arabic']) / 2
    student_profile['student_grades'] = {course : random.randint(60, 100) for course in student_profile['student_courses']}

    return student_profile


# Generate a dataset with N student profiles
def generate_dataset(n=10):
    return [generate_student_profile() for _ in range(n)]

# Generate the dataset
dataset = generate_dataset(100)

df_students = pd.DataFrame(dataset)
```

```python
from experta import *

class Level(Fact):
    student_level = Field(int, mandatory=True)

class TargetExam(Fact):
    tahsili = Field(bool)
    qudurat = Field(bool)

class Grades(Fact):
    math101 = Field(float)
    math102 = Field(float)
    math103 = Field(float)
    math201 = Field(float)
    math202 = Field(float)
    math203 = Field(float)
    math203 = Field(float)
    math301 = Field(float)
    math302 = Field(float)
    math303 = Field(float)

    physics102 = Field(float)
    physics203 = Field(float)
    physics301 = Field(float)
    physics302 = Field(float)
    physics303 = Field(float)

    bio101 = Field(float)
    bio201 = Field(float)
    bio202 = Field(float)
    bio203 = Field(float)

    chemistry103 = Field(float)
    chemistry201 = Field(float)
    chemistry202 = Field(float)
    chemistry203 = Field(float)
    chemistry301 = Field(float)

    arabic101 = Field(float)



class CourseRecommander(KnowledgeEngine):


    @Rule(AND(Level(student_level=MATCH.student_level)), TargetExam(tahsili=False, qudurat=True))
    def all_courses(self, student_level):
        print()
        print("RULE 1 IS FIRED")
        print("Based on your level, and your target exam is Qudurat, \nWe recommand you to study the following courses!")
        courses_level = df[df['course_level']>=student_level]
        courses = courses_level[courses_level['Qudurat?']==True]['course_name_en'].values
        print(courses)


    @Rule(AND(Level(student_level=P(lambda x: x < 7))), TargetExam(tahsili=True, qudurat=False))
    def tahsili_not_allowed(self):
        print("RULE 2 IS FIRED")
        print("It's great to be a hard worker! \nBut the Tahsili exam is not allowed for you! \nWait until you pass level 6, and then you can check again! \nAll best :)")

    @Rule(AND(Level(student_level=MATCH.student_level), TargetExam(tahsili=True, qudurat=True)))
    def tahsili_not_allowed_qudurat_allowed(self, student_level):
        if student_level < 7:  # Check if the student's level is greater than 6
            print()
            print("RULE 3 IS FIRED")
            print("It's great to be a hard worker!")
            print("But the Tahsili exam is not allowed for you!")
            print("Wait until you pass level 6, and then you can check again!")
            print("All the best :)")
            print("Based on your level and your target exam being Qudurat, we recommend you study the following courses:")

            # Filter courses based on the student's level
            courses_level = df[df['course_level'] >= student_level]
            courses = courses_level[courses_level['Qudurat?'] == True]['course_name_en'].values

            print("Recommended courses:")
            for course in courses:
                print(f"- {course}")

    @Rule(AND(
        Grades(
            math101=MATCH.math101_grade,
            math102=MATCH.math102_grade,
            math103=MATCH.math103_grade,
            math201=MATCH.math201_grade,
            math202=MATCH.math202_grade,
            math203=MATCH.math203_grade,
            math301=MATCH.math301_grade,
            math302=MATCH.math302_grade,
            math303=MATCH.math303_grade,
            physics102=MATCH.physics102_grade,
            physics203=MATCH.physics203_grade,
            physics301=MATCH.physics301_grade,
            physics302=MATCH.physics302_grade,
            physics303=MATCH.physics303_grade,
            bio101=MATCH.bio101_grade,
            bio201=MATCH.bio201_grade,
            bio202=MATCH.bio202_grade,
            bio203=MATCH.bio203_grade,
            chemistry103=MATCH.chemistry103_grade,
            chemistry201=MATCH.chemistry201_grade,
            chemistry202=MATCH.chemistry202_grade,
            chemistry203=MATCH.chemistry203_grade,
            chemistry301=MATCH.chemistry301_grade,
            arabic101=MATCH.arabic_grade
        ),
        TargetExam(qudurat=True)
    ))
    def Qudurat_allowed(self, math101_grade, math102_grade, math103_grade, math201_grade,math202_grade, math203_grade,
                         math301_grade, math302_grade, math303_grade, physics102_grade, physics203_grade,
                         physics301_grade, physics302_grade, physics303_grade, bio101_grade, bio201_grade,
                         bio202_grade, bio203_grade, chemistry103_grade, chemistry201_grade, chemistry202_grade,
                         chemistry203_grade, chemistry301_grade, arabic_grade):

        # List of all grades and their corresponding fields
        print()
        print("RULE 4 IS FIRED")
        grades = {
            "math101": math101_grade,
            "math102": math102_grade,
            "math103": math103_grade,
            "math201": math201_grade,
            "math202": math202_grade,
            "math203": math203_grade,
            "math301": math301_grade,
            "math302": math302_grade,
            "math303": math303_grade,
            "physics102": physics102_grade,
            "physics203": physics203_grade,
            "physics301": physics301_grade,
            "physics302": physics302_grade,
            "physics303": physics303_grade,
            "bio101": bio101_grade,
            "bio201": bio201_grade,
            "bio202": bio202_grade,
            "bio203": bio203_grade,
            "chemistry103": chemistry103_grade,
            "chemistry201": chemistry201_grade,
            "chemistry202": chemistry202_grade,
            "chemistry203": chemistry203_grade,
            "chemistry301": chemistry301_grade,
            "arabic101": arabic_grade
        }

        # Check each grade and print if below 90
        print("Based on you grades we recommend you to study and review the following courses: FOR QUDURAT EXAM!!")
        qudurat_courses = df[df['Qudurat?']==True]['course_id'].to_list()
        for course, grade in grades.items():
            if grade < 90 and course in qudurat_courses:  # Check if grade is numeric and less than 90

                print(f"{df[df['course_id']==course]['course_name_en'].values[0]}: The grade {grade} is below 90.")



    @Rule(AND(
        Grades(
            math101=MATCH.math101_grade,
            math102=MATCH.math102_grade,
            math103=MATCH.math103_grade,
            math201=MATCH.math201_grade,
            math202=MATCH.math202_grade,
            math203=MATCH.math203_grade,
            math301=MATCH.math301_grade,
            math302=MATCH.math302_grade,
            math303=MATCH.math303_grade,
            physics102=MATCH.physics102_grade,
            physics203=MATCH.physics203_grade,
            physics301=MATCH.physics301_grade,
            physics302=MATCH.physics302_grade,
            physics303=MATCH.physics303_grade,
            bio101=MATCH.bio101_grade,
            bio201=MATCH.bio201_grade,
            bio202=MATCH.bio202_grade,
            bio203=MATCH.bio203_grade,
            chemistry103=MATCH.chemistry103_grade,
            chemistry201=MATCH.chemistry201_grade,
            chemistry202=MATCH.chemistry202_grade,
            chemistry203=MATCH.chemistry203_grade,
            chemistry301=MATCH.chemistry301_grade,
            arabic101=MATCH.arabic_grade
        ),
        TargetExam(tahsili=True),
        Level(student_level=P(lambda x: x > 5))
    ))
    def Tahsili_allowed(self, math101_grade, math102_grade, math103_grade, math201_grade,math202_grade, math203_grade,
                         math301_grade, math302_grade, math303_grade, physics102_grade, physics203_grade,
                         physics301_grade, physics302_grade, physics303_grade, bio101_grade, bio201_grade,
                         bio202_grade, bio203_grade, chemistry103_grade, chemistry201_grade, chemistry202_grade,
                         chemistry203_grade, chemistry301_grade, arabic_grade):

        # List of all grades and their corresponding fields
        print()
        print("RULE 5 IS FIRED")
        grades = {
            "math101": math101_grade,
            "math102": math102_grade,
            "math103": math103_grade,
            "math201": math201_grade,
            "math202": math202_grade,
            "math203": math203_grade,
            "math301": math301_grade,
            "math302": math302_grade,
            "math303": math303_grade,
            "physics102": physics102_grade,
            "physics203": physics203_grade,
            "physics301": physics301_grade,
            "physics302": physics302_grade,
            "physics303": physics303_grade,
            "bio101": bio101_grade,
            "bio201": bio201_grade,
            "bio202": bio202_grade,
            "bio203": bio203_grade,
            "chemistry103": chemistry103_grade,
            "chemistry201": chemistry201_grade,
            "chemistry202": chemistry202_grade,
            "chemistry203": chemistry203_grade,
            "chemistry301": chemistry301_grade,
            "arabic101": arabic_grade
        }

        # Check each grade and print if below 90
        print("You are allowed to the Tahsili exam! \nBased on you grades we recommend you to study and review the following courses: ")
        for course, grade in grades.items():
            if grade < 90.0:  # Check if grade is numeric and less than 90
                print(f"{df[df['course_id']==course]['course_name_en'].values[0]}: The grade {grade} is below 90.")
```

```python
def chatbot():
    student_profile = {
        "student_level": 0,
        "Tahsili_target": 0,
        "Qudurat_target": 0,
        "student_courses": [],
        "previous_Qudurat_grade": {"math": 0, "arabic": 0, "all": 0},
        "previous_Tahsili_grade": 0,
        "student_grades": {}
    }

    # Introduction
    print("Welcome to the Student Profile Chatbot!")
    print("Please answer the following questions to build your profile.\n")

    # Step 1: Get Student Level
    while True:
        try:
            level = int(input("What is your student level (1-9)? "))
            if 1 <= level <= 9:
                student_profile["student_level"] = level
                break
            else:
                print("Please enter a number between 1 and 9.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # Step 2: Target Exams
    exams = input("Are you targeting the Tahsili exam, Qudurat exam, or both? (Type 'Tahsili', 'Qudurat', or 'both'): ").strip().lower()
    if exams == "tahsili":
        student_profile["Tahsili_target"] = 1
    elif exams == "qudurat":
        student_profile["Qudurat_target"] = 1
    elif exams == "both":
        student_profile["Tahsili_target"] = 1
        student_profile["Qudurat_target"] = 1
    else:
        print("Invalid input. Defaulting to 'none'. No target exams selected.")

    # Step 3: Previous Grades for Tahsili and Qudurat
    if student_profile["Tahsili_target"] and student_profile["student_level"] > 6:
        while True:
            try:
                tahsili_grade = float(input("What is your previous Tahsili grade? (Enter 0 if none): "))
                student_profile["previous_Tahsili_grade"] = tahsili_grade
                break
            except ValueError:
                print("Invalid input. Please enter a number.")

    if student_profile["Qudurat_target"]:
        while True:
            try:
                math_grade = float(input("What was your math grade in the previous Qudurat exam? (Enter 0 if none): "))
                arabic_grade = float(input("What was your Arabic grade in the previous Qudurat exam? (Enter 0 if none): "))
                all_grade = float(input("What was your overall grade in the previous Qudurat exam? (Enter 0 if none): "))
                student_profile["previous_Qudurat_grade"] = {"math": math_grade, "arabic": arabic_grade, "all": all_grade}
                break
            except ValueError:
                print("Invalid input. Please enter valid grades.")

    # Step 4: Student Courses (Based on Level)
    print("\nBased on your level, you need to enter your grades for relevant courses.")

    courses_by_level = df[df['course_level']<student_profile["student_level"]]['course_id'].to_list()[:-4]
    all_courses = df['course_id'].to_list()
    student_profile["student_courses"] = courses_by_level
    student_profile["student_grades"] = {course : 0.0 for course in all_courses}
    print(f"Relevant courses for your level: {', '.join(courses_by_level)}")
    for course in courses_by_level:
        while True:
            try:
                grade = float(input(f"Enter your grade for {course}: "))
                student_profile["student_grades"][course] = grade
                break
            except ValueError:
                print("Invalid input. Please enter a number.")

    # Final Summary
    print("\nYour student profile has been successfully built:")
    for key, value in student_profile.items():
        print(f"{key}: {value}")
    return student_profile

# Run the chatbot
student_profile = chatbot()
```

```python
engine = CourseRecommander()
engine.reset()
engine.declare(Level(student_level=student_profile['student_level']),
               TargetExam(tahsili=(student_profile['Tahsili_target']==1),
                          qudurat=(student_profile['Qudurat_target']==1)),
               Grades(**{field: student_profile['student_grades'][field] for field in student_profile['student_grades'].keys()})
              )

engine.run()
```
